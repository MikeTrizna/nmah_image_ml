```python
import pandas as pd
import imagehash
import pybktree
```

```python
hash_df = pd.read_csv('dupe_test_hash_results.tsv', sep='\t')
hash_df.info()
```

```python
hash_df = hash_df.dropna()
len(hash_df)
```

```python
import collections
Item = collections.namedtuple('Image',['image_hash', 'image_path'])
```

```python
hash_dict = hash_df[['image','dhash']].to_dict(orient='records')
hash_dict[:5]
```

```python
image_list = [Item(imagehash.hex_to_hash(image['dhash']), image['image']) for image in hash_dict]
```

```python
def image_diff(x, y):
    return(x.image_hash - y.image_hash)
```

```python
tree = pybktree.BKTree(image_diff, image_list)
```

```python
for image in image_list:
    hits = tree.find(image, 1)
    if len(hits) > 1:
        print(image.image_path)
        for hit in hits:
            score, image_match = hit
            if image_match.image_path != image.image_path:
                print('***',score, image_match.image_path)
```
